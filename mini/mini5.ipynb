```python
confusion_matrix = np.array([
  [65, 5, 3, 1, 1],
  [0, 54, 13, 2, 1],
  [1, 15, 33, 3, 2],
  [0, 2, 0, 78, 2],
  [1, 0, 6, 2, 57]
], dtype = np.int32)
disp = ConfusionMatrixDisplay(confusion_matrix)
disp.plot()
```

```python
for ix in range(confusion_matrix.shape[0]):
  cell = confusion_matrix[ix, ix]
  all = confusion_matrix[:, ix].sum() + confusion_matrix[ix, :].sum() - cell
  acc = (cell) / all
  print(ix, acc)
```

```python
for ix in range(confusion_matrix.shape[0]):
  cell = confusion_matrix[ix, ix]
  all = confusion_matrix[ix, :].sum()
  fp = confusion_matrix[ix, :].sum() - cell
  pre = cell / (all)
  print(ix, pre)
```
